## We import the libraries


```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

```

### We create the array of data

```python
price = np.array([10000, 150000, 180000, 210000, 240000, 270000, 300000])
print(price.ndim)

surface = np.array([50, 70, 80, 90, 100, 110, 120])
```

### We create and init the model

```python
model = LinearRegression()
```

### We fit the model

```python
model.fit(surface.reshape(-1, 1), price)
```

## Prepare the graphic

```python
earring = model.coef_
ordered_at_the_origin = model.intercept_

regression_surface = np.linspace(min(surface), max(surface), 100)


```

### We make a prediction


```python
# a = int(input("Enter the surface to predict the price: "))
# predict = model.predict([[a]])

# print(predict)

predict = earring * regression_surface +ordered_at_the_origin
```

```python
plt.scatter(surface, price , color="blue")
plt.plot(regression_surface, predict, color="red")
plt.xlabel("Superficie (metros cuadrados)")
plt.ylabel("Precio (euros)")
plt.show()
```
