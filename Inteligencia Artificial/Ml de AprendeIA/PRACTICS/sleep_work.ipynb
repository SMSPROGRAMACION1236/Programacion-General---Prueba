# What about
This program will try predicting the hours of sleep having the hours of work
## Steps

#### 1. Import All the libraries and load the data and put into train and test data

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from  sklearn.linear_model import LinearRegression
```

```python
## prepare the data
data = {'Hours of study': [2, 3, 4, 5, 6],
  'Hours of sleep': [5, 6, 7, 8, 9]}

df = pd.DataFrame(data)
print(df)
```

```python
## We prepare the train and test data
x = df['Hours of study'].values.reshape(-1, 1)
y = df['Hours of sleep'].values.reshape(-1, 1)
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

#### 2. Prepare the model


```python
## Init the model
lineal = LinearRegression()
```

```python
## We fit the model
lineal.fit(X_train, Y_train)
```

#### 3. We do the predict

```python
a = int(input("How long have you study: "))
y_pred = lineal.predict([[a]])
print(f"The person would sleep {y_pred[0][0]:.1f} hours after studying {a} hours .")
```

#### 4. Visualization

```python
plt.scatter(X_train, Y_train, color='blue', label='data_train')
plt.plot(X_train, lineal.predict(X_train), color='red', label='Lineal Regression')
plt.xlabel('Hours of Study')
plt.ylabel('Hours of Sleep')
plt.title(' Lineal Regression : Hours of sleep vs. Hours  of study')
plt.legend()
plt.show()
```
