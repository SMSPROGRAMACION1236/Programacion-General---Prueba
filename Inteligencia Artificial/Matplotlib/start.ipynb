```python
# We're going to import the libraries numpy and  pandas, which are required for this exercise.

import numpy as np
import matplotlib.pyplot as plt

X_data = np.random.random(500) * 100 # Random Data
y_data = np.random.random(500)* 100  # Random data
plt.scatter(X_data, y_data)
plt.show() # To show the data
```

# Giving Life, Decorate


```python
#To put colors
##We use the  c= as a argument to put a color
plt.scatter(X_data, y_data, c="violet") # we can also use hexadecimal values to express colors
plt.show()
```

```python
# To put markers, defaut is points
plt.scatter(X_data, y_data, c="violet", marker= "+", s= 123) # s= is to say the weight of the marker
plt.show()
```

```python
# Doing invisible or similar
plt.scatter(X_data, y_data, c="violet", marker= "+", s= 123, alpha= 0.09) # alpha in to modify the  transparency of the points
plt.show()
```

## Line Plots


```python
years = [2006  + x for x in range(16)]
weights = [80, 83, 84, 85, 86, 82, 81, 79, 83, 80, 82, 82, 83, 81, 80, 79]

plt.plot(years, weights, c= "red", lw = 2) # a line graphic, connecting the points, lw is how bigger is the line
plt.scatter(years, weights)# Just points
plt.show()
```

```python
plt.plot(years, weights, c= "red", lw = 2, linestyle ="--") # linestyle is how do you like to  see the line
```

## Bar plot

```python
x = ["python", "java", "php", "c#"]
y = [10,50, 30, 20]

plt.bar(x, y, color= "red", align="edge", width= 0.5, edgecolor="green", lw= 10) # We need to say color and not c= and align edge is the  default so we can omit it, and width is the the weight of the bars and edgecolor the color of the edge
```

## Histograms

```python
ages = np.random.normal(20, 1.5, 1000) # The mean is 20, the standard deviation is 1.5 and 1000  random numbers are generated


plt.hist(ages, bins=20, cumulative= True) # bins is similar like the weight and cumulative is like to  increase it
```

```python
langs = ['Python', "c++", "java", "c#",  'JavaScript']
votes = [20, 30, 45, 15, 6]
explodes = [0.3, 0, 0, 0, 0 ]
plt.pie(votes, labels= langs, explode= explodes) # The labels to identify the values explode to separate data from another one is like taking out a piece of the cake
```

```python
# Put percent

plt.pie(votes, labels= langs, explode= explodes, autopct="%.2f%%", pctdistance= 1.4) # autopct is to put the percent and pctdistance is to put the distance between those ones
```

## Box Plots

```python
height = np.random.normal(172, 8, 300) # # The mean is 172, the standard deviation is 8 and 300  random numbers are generated
plt.boxplot(height)
```

```python
first = np.linspace(0, 10, 25)
second = np.linspace(10, 200,25)
third = np.linspace(200, 210, 25)
fourth = np.linspace(210, 230, 25)

data = np.concatenate((first, second, third, fourth))

plt.boxplot(data)
```
