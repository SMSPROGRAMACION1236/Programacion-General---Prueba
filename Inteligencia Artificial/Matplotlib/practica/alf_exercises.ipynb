#### Ejercicio 1
Escribir un programa que pregunte al usuario por las ventas de un rango de años y muestre por pantalla un diagrama de líneas con la evolución de las ventas.

```python
year =  int(input("Enter the range of years: "))
years = []
sells = []

for i in range(1, year +1):
  years.append(i)
  
for i in range(year):
  sells.append(int(input("Enter the number of sells: ")))


plt.plot(years,sells, color="red")
plt.xlabel("Years")
plt.ylabel("Sells")
plt.title("Sells in a range of years")
plt.show()
  
```

#### Ejercicio 2
Escribir una función que reciba una diccionario con las notas de las asignaturas de un curso y una cadena con el nombre de un color y devuelva un diagrama de barras de las notas en el color dado.




```python
def dict_to_plot(both :dict):
  try:
    color = str(input("What color do you like to be the bars: "))
  except ValueError:
    print("Please enter a valid color")
  else:
    plt.bar(list(both.keys()),list(both.values()),color=color)
    return plt.show()

dict_to_plot({"math":23,"literature":12})

```

#### Ejercicio 3
Escribir una función que reciba una serie de Pandas con las notas de los alumnos de un curso y devuelva un diagrama de cajas con las notas. El diagrama debe tener el título “Distribución de notas”.


```python

def box_plot(box):
  plt.boxplot(box)
  plt.title("Distribucion de Notas")
  plt.show()
print(box_plot(pd.Series([2,5])))
```
