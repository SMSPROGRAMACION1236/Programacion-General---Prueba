```python
import matplotlib.pyplot as plt 
```

#### Grafico de  linea
Crear un gráfico de línea simple con los siguientes datos:

Eje X: 1, 2, 3, 4, 5

Eje Y: 2, 4, 5, 4, 3


```python
eje_X= [1, 2, 3, 4, 5]
eje_Y= [2, 4, 5, 4, 3]
plt.plot(eje_X, eje_Y,color = "red")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

#### grafic de barras
Crear un gráfico de barras con los siguientes datos:

Ciudades:

Madrid, Barcelona, Valencia, Sevilla

Población: 

3.2 millones, 5.5 millones, 800 mil, 1.7 millones

```python
ciudades = ["Madrid","Barcelona","Valencia","Sevilla"]
poblacion = [3.2, 5.5 ,0.8, 1.7  ]

plt.bar(ciudades, poblacion,color = "red", edgecolor= "green")
plt.xlabel("ciudades")
plt.ylabel("Cantidad(en millones)")
plt.title("Poblacion de ciudades")
plt.show()
```

#### Grafico de Pastel
Crear un gráfico de pastel con los siguientes datos:

Sectores: Transporte, Vivienda, Alimentación, Otros

Porcentajes: 30%, 40%, 20%, 10%


```python
sectores =["transporte","vivienda","alimentacion","otros"]
porcentaje = [30, 40, 20, 10]
plt.pie(porcentaje,labels=sectores,autopct= "%.2f%%")
```
